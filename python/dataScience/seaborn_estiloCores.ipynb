```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

- Modficando o estilo dos graficos com sns e plt
- sns.despine() tira as borda

```python
plt.figure(figsize=[4,2])
sns.set_style(style='whitegrid')
sns.countplot(data=tips,x ='sex')
```

- Aspect e height(size)

```python
sns.lmplot(data=tips,x='total_bill',y='tip',aspect=2,height=4)
```

- sns.set_context(): alterar o contexto no qual ografico gerado ser√° usado. alterar o tamanho da fonte
- pallete: alterar a plalheta de cores

```python
sns.set_context('poster',font_scale=1)
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',palette='plasma')
```
