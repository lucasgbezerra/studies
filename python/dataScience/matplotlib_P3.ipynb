```python
import matplotlib.pyplot as plt
import numpy as np
```

### Layouts de matplotlib

```python
x = np.linspace(0,5,11)
```

```python
y= x*x
```

- Resumo matplotlib tem mais opções de design dos graficos

```python
fig,axes = plt.subplots(figsize= (10,10))
axes.plot(x,x**2,color='purple', linewidth= 5, alpha= 0.5, linestyle= '--',marker ='o', markersize=10, markerfacecolor='green')
axes.plot(y,x,color='green', linewidth= 2, linestyle= '-',marker ='s', markersize=10, markerfacecolor='black')
axes.set_title('Titulo')
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

-  Criando um limitador nos eixos, dá uma espécie de zoom nos dados

```python
axes.set_xlim([0,5])
axes.set_ylim([0,5])

fig
```

### Tipos especiais de plot

```python
plt.scatter(x,y)
plt.show()
```

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# box plot retangular
plt.boxplot(data,vert=True,patch_artist=True);   
```
