### Plots categóricos
- Plots utilizados para analisar os dados não númericos
- barplot(): apresenta um grafico em barra que cruza dados númericos e não númericos
- countplot(): reperesentação da quantidade de dados não numéricos
- boxplot():As hastes inferiores e superiores se estendem, respectivamente, do quartil inferior até o menor valor não inferior ao limite inferior e do quartil superior até o maior valor não superior ao limite superior
- violinplot(): representação dos dados por KDE
- sns.stripplot(): plota pontos onde os dados aparecem
- sns.swarmplot(): coloca ao lado os dados conforme aparecem formando uma especie de enxame
- factorplot(): 

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

- no **barplot()** o traço representa o desvio padrao, e a altura no caso acima a média

- **countplot()**, faz, basicamente, a representação gráfica da contagem do dado não numérico 

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

- É possível utilizar o **hue**
- É possível alterar a orientação
- a linha abaixo gera um boxplot de todos os dados numéricos

```python
sns.boxplot(data=tips,orient='h')
```

```python
sns.violinplot(data=tips,x='day',y='tip')
```

-  o ponto representa a média
- a representação é simétrica o que torna usar o hue da forma abaixo atrativo

```python
sns.violinplot(data=tips,x='day',y='tip',hue='sex',split=True)
```

```python
sns.stripplot(data=tips, x='day', y='total_bill',jitter)
```

- Tira a noção de densidade dos dados
- jitter como falso aglomera os dados em linha
- com o hue o split como True separa os dados do hue

```python
sns.stripplot(data=tips, x='day', y='total_bill',hue='sex',split=True)
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

- Pode ser problemático com quantidade muito grande de dados podendo até sobrepor as colunas

```python
sns.swarmplot(data=tips, x='day', y='total_bill',hue='sex',split=True)
```

### Mesclando gráficos
- Pode ser interessante mesclar representações

```python
sns.swarmplot(data=tips, x='day', y='total_bill',color='black')
sns.violinplot(data=tips,x='day',y='total_bill')
```
