```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

- plot de Regressão
- modelo linear
- permite a predição do comportamento


```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'], col='sex',row='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'], col='sex',row='smoker', scatter_kws={'s':100})
```

- Alterar aspectos do gráfico

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', col='day')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', col='day',aspect=0.5,size=8)
```
