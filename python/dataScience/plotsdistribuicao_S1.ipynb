```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips
```

```python
tips.info()
```

- Distplot padrão

```python
sns.distplot(tips['total_bill'])
```

- Distplot sem o kde(kernel density estimation)
- alterando o bins

```python
sns.distplot(tips['total_bill'], kde=False, bins=50)
```

- Jointplot: graficos de distribuição conjunta

```python
sns.jointplot(data =tips, x='total_bill',y='tip')
```

- Uso do kind para modificação do plot a ser desenhado,  reg = regressaão linear, hex hexagonos de 'calor'

```python
sns.jointplot(data =tips, x='total_bill',y='tip',kind='reg' )
```

-  pairplot, faz o plot de todos os dados númericos do DF

```python
sns.pairplot(data=tips)
```

- hue: parametro que adiciona mais informações aos graficos com dados como fumante ou não e sex 

```python
tips.head()
```

```python
sns.pairplot(data=tips,hue='sex')
```

### Distribuição Normal e KDE
- A distribuição é normal quando tem a forma de "sino"
- Em Estatística, estimativa de densidade por Kernel (EDK) é uma forma não-paramétrica para estimar a Função densidade de probabilidade (FDP) de uma variável aleatória. Estimativa da densidade por Kernel é um problema fundamental de suavização de dados onde inferências sobre a população são feitas com base em uma amostra de dados finita

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
