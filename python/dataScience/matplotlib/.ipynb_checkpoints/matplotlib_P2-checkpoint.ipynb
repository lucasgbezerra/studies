```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,5)
```

```python
x
```

```python
y = x*x
```

```python
y
```

- subplots, trabalha com a instancia da clase

```python
fig, ax = plt.subplots()
```

```python
ax.plot(x,x*x,'b-.')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('1 Titulo')
fig
```

```python
fig, ax = plt.subplots(nrows= 1, ncols=2)
plt.tight_layout()
```

- **ax** é uma array de duas posiçoes, onde cada uma é uma das figuras

```python
ax
```

```python
ax[0].plot(x,x**2,'b-.')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_title('1 Titulo')

ax[1].plot(x,x**5,'r--')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Z')
ax[1].set_title('2 Titulo')
fig
```

- altera o tamanho da figura, dpi é a resolução onde 100 é o padrão

```python
fig = plt.figure(figsize=(8,8),dpi=100)
```

- Criando um gráfico com legenda posicionada pelo **loc()**, padrão é o loc(0), seleciona o local com menos informações
- Alterando as configurações de tamanho da figura

```python
fig, axes = plt.subplots(figsize=(8,8))

axes.plot(x,y,'r--', label='x²')
axes.plot(y,x,'g*-', label = 'sqrt(x)')
axes.set_title('Titulo')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.legend()
```
