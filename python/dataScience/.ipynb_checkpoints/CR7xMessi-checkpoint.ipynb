```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
messiDf = pd.read_excel('dadosm.xlsx')
cr7Df = pd.read_excel('dadoscr7.xlsx')
```

```python
messiDf
```

```python
messiDf.fillna(0,inplace=True)
cr7Df.fillna(0,inplace=True)
```

```python
messiDf.head()
```

```python
df10 = messiDf.drop([16], axis=0)
df10
```

```python
cr7Df
```

```python
df7 = cr7Df.drop([0,1,18],axis=0)
```

```python
fig, ax= plt.subplots(figsize=(20,10))
ax.plot(df10['Temporada'],df10['Total - golos'],'r-',marker='o',label='Messi')
ax.plot(df7['Temporada'],df7['Total - golos'],'g--',marker='*',label='CR7')
ax.legend()
ax.grid(True)
```

```python
fig, ax= plt.subplots(figsize=(20,10))
ax.plot(df10['Temporada'],df10['Soma - gols'],'r-',marker='o')
ax.plot(df7['Temporada'],df7['Soma - gols'],'g-',marker='o')
ax.grid(True)
```
