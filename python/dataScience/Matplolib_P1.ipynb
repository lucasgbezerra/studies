```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

- Manneira mais simples de plotar graficos de maneira funcional
- plt.show(), só é necessario caso o matplotlib inline não seja utilizado

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color= 'r')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
plt.show()
```

- Segunda forma de plot
- plt.subplot(nrows =1, ncols =2,index =1)

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
plt.show()
```

- Forma mais comum de se plotar gráficos
- Oferece maior controle e liberdade sobre o grafico criado.
- Envolve OO
- **plt.figure()** cria uma "caixa vazia", que é um objeto com instancias possiveis de preenchimento

```python
fig = plt.figure()
axs1 = fig.add_axes([0.1,0.1,0.8,0.8])
axs1.plot(x,y,'r--')
axs1.set_title('Título')
axs1.set_xlabel('Eixo X')
axs1.set_ylabel('Eixo Y')
axs2 = fig.add_axes([0.2,0.4,0.4,0.4])
axs2.plot(y,x,'b*-')
axs2.set_title('Título 2')
axs2.set_xlabel('Eixo X')
axs2.set_ylabel('Eixo Y')
```
